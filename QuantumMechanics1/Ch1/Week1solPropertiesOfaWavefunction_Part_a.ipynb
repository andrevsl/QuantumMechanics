```python
%%latex
$
or \\

N^2\intop_{ -\infty }^{\infty }  x^2e^{(-  \alpha  x^2)   }     dx =\\

integrating \ by \ parts,


since \ \int u^{'}v dx=uv -   \int v^{'}udx                   \\
 
 
imposing \ u=-e^{(-  \alpha  x^2)}/(2\alpha) \ and \ v=x  \\


N^2\intop_{ -\infty }^{\infty }  x^2e^{(-  \alpha  x^2)   } dx=  \\
 
 -xe^{(-  \alpha  x^2)}/(2\alpha) -  \intop_{ -\infty }^{\infty } (-e^{(-  \alpha  x^2)}/(2\alpha))dx
=\frac{1 }{N^2}\\
 
 
  -xe^{(-  \alpha  x^2)}/(2\alpha) +  \intop_{ -\infty }^{\infty } e^{(-  \alpha  x^2)}/(2\alpha)dx
=\frac{1 }{N^2}\\

 
 -xe^{(-  \alpha  x^2)}/(2\alpha) +  \intop_{ -\infty }^{0 } e^{(-  \alpha  x^2)}/(2\alpha) dx
 +   \intop_{ 0 }^{\infty } e^{(-  \alpha  x^2)}/(2\alpha)dx =\frac{1 }{N^2}\\
 
 
 -xe^{(-  \alpha  x^2)}/(2\alpha) -  \intop_{ 0 }^{-\infty} e^{(-  \alpha  x^2)}/(2\alpha) dx
 +   \intop_{ 0 }^{\infty } e^{(-  \alpha  x^2)}/(2\alpha)dx =\frac{1 }{N^2}\\
 
 being  \ t^2=\alpha x^2  \ \rightarrow  t= \sqrt{\alpha } x   \ \rightarrow \
 
 \frac{\mathrm{d} t }{\mathrm{d} x} =\sqrt{\alpha }  \rightarrow  =\frac{1 }{N^2} \\ 
 
 \frac{\mathrm{d} t }{\sqrt{\alpha }} =\mathrm{d} x =\frac{1 }{N^2} \\
 
 
  -xe^{(-  \alpha  x^2)}/(2\alpha) {|_{-\infty} ^{+\infty}} 
  - \frac{1}{2\alpha ^{(3/2)}}\intop_{ 0 }^{-\infty} e^{(-  t^2)} dt
 +   \frac{1}{2\alpha ^{(3/2)}}\intop_{ 0 }^{\infty } e^{(-   t^2)}dt =frac{1 }{N^2}\\
 
 
   -xe^{(-  \alpha  x^2)}/(2\alpha) {|_{-\infty} ^{+\infty}} 
  - \frac{\sqrt{\pi}}{4\alpha ^{(3/2)}} \frac{2 }{\sqrt{\pi }}  \intop_{ 0 }^{-\infty} e^{(-  t^2)}dt
 +   \frac{\sqrt{\pi}}{4\alpha ^{(3/2)}} \frac{2 }{\sqrt{\pi }} \intop_{ 0 }^{\infty } e^{(-   t^2)}dt
=\frac{1 }{N^2}\\
 
 
   -xe^{(-  \alpha  x^2)}/(2\alpha) {|_{-\infty} ^{+\infty}} 
  - \frac{\sqrt{\pi}}{4\alpha ^{(3/2)}} \frac{2 }{\sqrt{\pi }}  \lim_{t\to -\infty}  erf(t)
 +   \frac{\sqrt{\pi}}{4\alpha ^{(3/2)}} \frac{2 }{\sqrt{\pi }} \lim_{t\to \infty}  erf(t)
=\frac{1 }{N^2}\\
 
 
 
 since \ the \ first \ term \ vanishes for +/- \infty

 \\
 \frac{\sqrt{ \pi } \ erf( \sqrt{ \alpha  } x)}{4  \alpha^{3/2} }   {|_{-\infty} ^{+\infty}}   
    =\frac{1 }{N^2}\\
 \frac{N^2\sqrt{ \pi } }{2  \alpha^{3/2} } =1   \\
    

N^2  =2\sqrt{\frac{ \alpha ^3   }{ \pi  }} \\

N   =  \sqrt[4]{\frac{ 4\alpha ^3   }{\pi  }} \\

$
```
