```python
import numpy as np
from sympy import *
def calcCn(N):
    n=symbols('n')
    f=sqrt(3)/pi*(sin((n-3)*pi/3)/(n-3)-sin((n+3)*pi/3)/(n+3))
    y=limit(f,n,N)
    return round(y,4)

def enStates(N):
    hbar=symbols('hbar')
    m=symbols('m')
    L=symbols('L')
    p=symbols('p')
    f=(N**2*p**2*hbar**2/(18*m*L)).evalf()
    return f
#1=calcCn(1)
print(calcCn(1))

print(calcCn(2))
print(calcCn(4))
print(calcCn(5))
print(calcCn(6))
print(enStates(3))
```
